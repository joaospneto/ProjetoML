```python
import yfinance as yf
```

```python
ticker = "AAPL"  # Replace with the desired ticker symbol
```

```python
dados = yf.Ticker(ticker).history("3y")
```

```python
dados.head()
```

```python
treinamento = dados.reset_index()
```

```python
treinamento["Date"] = treinamento["Date"].dt.date
treinamento = treinamento[['Date', 'Close']]
treinamento.columns = ['ds', 'y']
treinamento
```

```python
from prophet import Prophet
```

```python
modelo = Prophet() # Cria um novo objeto Prophet

```

```python
modelo.fit(treinamento) # Ajusta o novo modelo com os dados
```

```python
datas_futuras = modelo.make_future_dataframe(30)
datas_futuras
```

```python
previsoes = modelo.predict(datas_futuras)
previsoes.head()
```

```python
from prophet.plot import plot_plotly
```

```python
plot_plotly(modelo, previsoes)
```
